```python
outlier=data[data.price>50][['track_name','price','prime_genre','user_rating']]
outlier

paidapps=data[(data['price']<50) & (data['price']>0)]

plt.figure(figsize=(10,5))
plt.hist(paidapps['price'],log=True)
plt.title('Price distribution of apps (Log scale)')
plt.ylabel('Frequency Log scale')
plt.xlabel('Price Distributions in ($)')

# Count of paid apps exponentially decreases as the price increases.
# Very few apps are priced above $30.
```

* Count of paid apps is exponentially decreases as the price increases.
* Very few apps have been priced above 30, so it's important to keep price of your app below 30.

```python
plt.figure(figsize=(10,5))
plt.hist(data['user_rating'],bins=5)
plt.title('Rating')
plt.ylabel('Frequency')
plt.xlabel('Raing Distributions')

# Most of apps are rated 4.5 and very few are rated as average 5.
```

* Most of apps are rated 4.5 and very few are rated as average 5.

```python
# Create a new column based on the price of each app
data['category'] = np.where(data['price']!=0.00, 'paid', 'free')

# Create a new column based on the prime genre of each app
s = data['prime_genre'].value_counts().index[:4]
def categ(x):
    if x in s:
        return x
    else : 
        return "Others"
data['broad_genre']= data['prime_genre'].apply(lambda x : categ(x))

# Create a new column based on the size of each app
data['size_MB'] = data['size_bytes']/(1024*1024)
```

### 1.4 What is the price distribution over prime genre?

```python
# The 4 most popular categories are Games, Entertainment, Education, and Photo & Video.
price_Games=data[data['prime_genre']=='Games']['price']
price_Entertainment=data[data['prime_genre']=='Entertainment']['price']
price_Education=data[data['prime_genre']=='Education']['price']
price_Phovid=data[data['prime_genre']=='Photo & Video']['price']
```

```python
fig,ax=plt.subplots(nrows=1,ncols=4,figsize=(20,10))

bplot1=ax[0].boxplot(price_Games,notch=True,vert=True,patch_artist=True)
ax[0].set_xlabel('Games')
ax[0].set_ylabel('Price distribution')
for box in bplot1['boxes']:
    box.set(color='pink')
    
bplot2=ax[1].boxplot(price_Entertainment,notch=True,vert=True,patch_artist=True)
ax[1].set_xlabel('Entertainment')
ax[1].set_ylabel('Price distribution')
for box in bplot2['boxes']:
    box.set(color='tomato')
    
bplot3=ax[2].boxplot(price_Education,notch=True,vert=True,patch_artist=True)
ax[2].set_xlabel('Education')
ax[2].set_ylabel('Price distribution')
for box in bplot3['boxes']:
    box.set(color='lightgreen')

bplot4=ax[3].boxplot(price_Phovid,notch=True,vert=True,patch_artist=True)
ax[3].set_xlabel('Photo & Video')
ax[3].set_ylabel('Price distribution')
for box in bplot4['boxes']:
    box.set(color='lightblue')

# Paid gaming apps are highly priced and the distribution extends till $30.
# Paid entertainment apps have a lower price range.
# There are some high-price apps that exceed $50 in education.
```

### Insights:
* Paid gaming apps are highly priced and the distribution extends till 30.
* Paid entertainment apps have a lower price range.
* There are some high-price apps that exceed 50 in education.

## 2. Number of Devices Exploration

### 2.1 Is there any relationship between rating and number of devices supported ?

```python
data['sup_devices.num'].corr(data['user_rating'])

sup_devices_ratingcount=data.groupby('sup_devices.num').count()['user_rating']
sup_devices_ratingaverage=data.groupby('sup_devices.num').mean()['user_rating']
sup_devices_number=data['sup_devices.num'].sort_values().unique()

sup_devices_ratingcount
sup_devices_ratingaverage
sup_devices_number

# The relationship between average rating and number of devices supported is -0.04245, which is not evident.
# 37 supported devices gets the most ratings and a relatively high average rating, so it may be the optimal number.
```

```python
fig,ax=plt.subplots(figsize=(20,10))
ax.plot(sup_devices_number,sup_devices_ratingaverage,
        color="violet", lw=1, ls='-', 
        marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
ax.set_xlabel('Number of devices supported')
ax.set_ylabel('Average rating')
ax.set_title('Relationship Between Number of Devices Supported and Average Rating')


fig,ax=plt.subplots(figsize=(20,10))
ax.plot(sup_devices_number,sup_devices_ratingcount,
        color="violet", lw=1, ls='-', 
        marker='s', markersize=8, markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green")
ax.set_xlabel('Number of devices supported')
ax.set_ylabel('Number of rating')
ax.set_title('Relationship Between Number of Devices Supported and Number of Rating')
```

### Insights:
* The relationship between average rating and number of devices supported is -0.04245, which is not evident.
* 37 supported devices gets the most ratings and a relatively high average rating, so it may be the optimal number.
